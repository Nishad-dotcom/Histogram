#Seaborn


```python
import seaborn as sns
```

```python
# loading data
data=sns.load_dataset('iris')
```

```python
data.head(2)
```

```python
# histogram 
sns.histplot(data=data,x='petal_length',kde=True)
plt.xlabel('petal length')
plt.ylabel('frequency')
plt.title('Histogram of petals')
plt.show()
```

# Box plot

```python
# sample data
df=sns.load_dataset('tips')
```

```python
df.head(1)
```

```python
# create a box plot
sns.boxplot(data=df,x='day',y='total_bill',palette='rocket')
# add label
plt.xlabel('Day')
plt.ylabel('Total bill')
plt.title('Box plot')
plt.show()
```

# voilin plot

```python
sns.violinplot(data=df,x='day',y='total_bill',palette='rocket')
# add label
plt.xlabel('Day')
plt.ylabel('Total bill')
plt.title('violin plot')
plt.show()
```
